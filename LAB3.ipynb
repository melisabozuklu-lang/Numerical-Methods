```python
# Function definition
def f(x):
    return x*x - S

def g(x):
    return 2*x

# Newton / Babylonian Method
def newton(x0, e, N):
    print("\n*** NEWTON METHOD (BABYLONIAN METHOD) ***\n")

    step = 1
    flag = 1
    condition = True

    while condition:
        if g(x0) == 0.0:
            print("Divide by zero error!")
            break

        x1 = x0 - f(x0)/g(x0)

        print("Iteration-%d, x1 = %0.10f" % (step, x1))

        x0 = x1
        step = step + 1

        if step > N:
            flag = 0
            break

        condition = abs(f(x1)) > e

    # result
    if flag == 1:
        print("\nRequired root is: %0.10f" % x1)

        real = S**0.5
        if x1 > real:
            print("Overestimate")
        elif x1 < real:
            print("Underestimate")
        else:
            print("Exact result")

    else:
        print("\nNot Convergent.")

# Input Section
S = float(input("Enter S value: "))
x0 = float(input("Enter initial guess: "))
e = float(input("Enter tolerable error: "))
N = int(input("Enter maximum steps: "))

# Starting Newton Method
newton(x0, e, N)
```
