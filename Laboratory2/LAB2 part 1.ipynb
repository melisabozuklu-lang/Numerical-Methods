```python
import numpy as np

def test_precision(dtype):
    print(f"\nTesting: {dtype}")
    n = 0

    while True:
        eps = dtype(2.0) ** dtype(-n)
        x = dtype(1.0) + eps

        print(f"n = {n}, epsilon = {eps}, fl(1+eps) = {x}")

        if x == dtype(1.0):
            print(f"1 + epsilon becomes 1 at n = {n}")
            break

        n += 1

test_precision(np.float16)
test_precision(np.float32)
test_precision(np.float64)

if hasattr(np, "float128"):
    test_precision(np.float128)
```
