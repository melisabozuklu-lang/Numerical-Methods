```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x * np.sin(4*x)

def g(x):
    return np.sin(4*x) + 4*x*np.cos(4*x)

def newton(x0, TOLX, TOLF, N):
    for step in range(1, N + 1):

        if abs(g(x0)) < 1e-12:
            return None, step

        x1 = x0 - f(x0) / g(x0)

        error_x = abs(x1 - x0)
        error_f = abs(f(x1))

        if error_x <= TOLX and error_f <= TOLF:
            return x1, step

        x0 = x1

    return x1, step


TOLX = 0.00001
TOLF = 0.00001
N = 20

initial_guesses = np.arange(-2, 2.1, 0.25)

roots = []
iterations = []

print("Initial Guess\tRoot\t\tIterations")
print("----------------------------------------")

for x0 in initial_guesses:
    root, it = newton(x0, TOLX, TOLF, N)

    roots.append(root)
    iterations.append(it)

    if root is not None:
        print(f"{x0:.2f}\t\t{root:.6f}\t{it}")
    else:
        print(f"{x0:.2f}\t\tNo convergence\t{it}")


# Figure 1: Function only, like teacher's first graph
x = np.linspace(-2, 2, 1000)

plt.figure()
plt.plot(x, f(x), label="function")
plt.axhline(0, color="orange")
plt.title("f(x) = x sin(4x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid()
plt.legend()
plt.show()


# Figure 2: Initial guess vs iteration number
plt.figure()
plt.plot(initial_guesses, iterations, marker="o")
plt.title("Initial Guess vs Number of Iterations")
plt.xlabel("Initial Guess")
plt.ylabel("Number of Iterations")
plt.grid()
plt.show()
```
